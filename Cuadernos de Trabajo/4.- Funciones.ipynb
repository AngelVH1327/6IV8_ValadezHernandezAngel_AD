## 4.3.- Funciones con M煤ltiples Par谩metros

Ahora armaremos una funci贸n usando IF que nos dice cual es la raiz cuadrada m谩s grande de 3 n煤meros

```python
def raiz(num1,num2,num3):
    r_num1 = num1**(1/2)
    r_num2 = num2**(1/2)
    r_num3 = num3**(1/2)
    raiz = r_num1
    if r_num2 > r_num1:
        raiz = r_num2
    if r_num3 > r_num2:
        raiz = r_num3
    return raiz

raiz(4,25,9)
```

Por si las dudas, esto se podr铆a haber simplificado usando Max y Min

```python
def raiz(num1, num2, num3):
    return max(num1**(1/2), num2**(1/2), num3**(1/2))

raiz(4, 25, 9)
```

Ahora, vamos  a simular las consecuencias del primer tiro - si la suma es 8, ganaste, si es 7, perdiste, y si no, continuamos

```python
def lanzar_dados():
    dado1 = random.randrange(1, 7)
    dado2 = random.randrange(1, 7)
    return(dado1,dado2)

def mostrar_dados(dado):
    dado1,dado2 = dado
    print(f'Resultado:  {dado1} + {dado2} = {sum(dado)}')

valores_logrados = lanzar_dados()

mostrar_dados(valores_logrados)

suma_dados = sum(valores_logrados)

if suma_dados == 8:
    print("Ganaste")
elif suma_dados == 7:
    print("Perdiste")
else:
    print("El juego contin煤a")
```

Y ahora con un bucle vamos a seguir el juego, se relanzan los dados y si se hace un bucle while que mientras que no haya un resultado de 6,8, o 7, se sigue tirando, si no se define si se gan贸 o se perdi贸

```python
def lanzar_dados():
    dado1 = random.randrange(1, 7)
    dado2 = random.randrange(1, 7)
    return(dado1,dado2)

def mostrar_dados(dado):
    dado1,dado2 = dado
    print(f'Resultado:  {dado1} + {dado2} = {sum(dado)}')

valores_logrados = lanzar_dados()

mostrar_dados(valores_logrados)

suma_dados = sum(valores_logrados)

if suma_dados == 8:
    status = 'ganaste'
elif suma_dados == 7:
    status = 'perdiste'
else:
    status = 'continuar'
    print('sigue probando')

while status == 'continuar':
    valores_logrados = lanzar_dados()
    mostrar_dados(valores_logrados)
    suma_dados = sum(valores_logrados)
    if suma_dados in (6,8):
        status = 'ganaste'
    elif suma_dados == 7:
        status = 'perdiste'

if status == 'ganaste':
    print('ganaste')
else:
    print('perdiste')
```

## 4.8.- Medidas de tendencia central y Dispersi贸n

Importa la librer铆a d eestadistica, y crea una lista de calificaciones aleatorias entre 50 y 100

```python
import statistics as estad
import random

calificaciones = [random.randrange(50, 101) for numero in range(10)]

print("Calificaciones:", calificaciones)
```

Calcula las medidas de tendencia central

```python
media = estad.mean(calificaciones)
mediana = estad.median(calificaciones)
moda = estad.mode(calificaciones)

print("\nLas medidas de tendencia central son:")
print("Media:", media)
print("Mediana:", mediana)
print("Moda:", moda)
```

Calcula las medidas de dispersi贸n

```python
rango = max(calificaciones) - min(calificaciones)
varianza = estad.pvariance(calificaciones)
desv_estandar = estad.pstdev(calificaciones)

print("\nLas medidas de dispersi贸n son:")
print("Rango:", rango)
print("Varianza:", round(varianza, 3))
print("Desv. Est谩ndar:", round(desv_estandar, 3))
```
